```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/stud.csv')
```

```python
df.info()
```

```python
print(df['gender'].unique())
print(df['race_ethnicity'].unique())
print(df['parental_level_of_education'].unique())
print(df['lunch'].unique())
print(df['test_preparation_course'].unique())
```

```python
df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']
```

```python
gender_mean_score = df.groupby('gender')['total_score'].describe()
ethnicity_mean_score = df.groupby('race_ethnicity')['total_score'].describe()
parent_mean_score = df.groupby('parental_level_of_education')['total_score'].describe()
lunch_mean_score = df.groupby('lunch')['total_score'].describe()
test_mean_score = df.groupby('test_preparation_course')['total_score'].describe()
print(gender_mean_score)
print(ethnicity_mean_score)
print(parent_mean_score)
print(lunch_mean_score)
print(test_mean_score)
```
